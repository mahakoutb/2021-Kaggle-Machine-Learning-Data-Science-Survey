# Data Cleaning
### Here we will start to clean our data 

```python
#As we saw the first raw have the questions of survey that we shouldn't be included in our analysis,
#so i will save it in 'Queastions' then drop it from dataset

#new data frame for 1st row 
Questions = df.iloc[0]
```

```python
Questions
```

```python
#drop first row 

df.drop(df.index[0], inplace = True)
df.head(2)
```

```python
#Now We have to check if all values in data set are objects and in case there is numeric values so we should fix this problm

#here we choose the column that displayed to check if it has a numeric values
df['Time from Start to Finish (seconds)'].str.isnumeric().all()
```

```python
# here we convert numeric string to numeric values 

for col in df.columns:
    if df[col].str.isnumeric().all():
        df[col]=pd.to_numeric(df[col])
    

    
```

```python
#check data types after fix the problem 

df.info()
```

```python
#Create Dictionary contains counts of each programing langauge 

dic_7 = dict()
for col in Q7_columns :
    key = df_top_countries[col].value_counts().index[0]
    dic_7[key] = df[col].value_counts()[0]
dic_7    
```

```python
#convert Dictionary to series

Q7_Series = pd.Series(dic_7)
Q7_Series
```

```python
Q7_Series = Q7_Series.sort_values(ascending = False)
```

```python
plt.figure(figsize = (10,10))

plt.pie(Q7_Series, labels =Q7_Series.index , autopct = '%1.1f%%')
plt.legend(Q7_Series.index)
plt.show()
```

```python
Cust_color = ['#6ee1fa', '#65d4ee']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20,10))
fig.patch.set_facecolor('#F2F2F2')


axes[0].bar(Q7_Series.index , Q7_Series.values, color =Cust_color  )
axes[0].set_ylabel('count' , weight= 'semibold' , fontname ='monoscape' , size = 20)
axes[0].set_xlabel('Programing Language' , weight= 'semibold' , fontname ='monoscape' , size = 20)
axes[0].tick_params(labelrotation = 80 , labelsize = 15)
fig.patch.set_facecolor('#F2F2F2')


axes[1].pie(Q7_Series, labels =Q7_Series.index , autopct = '%1.1f%%', colors = Cust_color , explode = [.04 for i in range(len(Q7_Series))])
axes[1].set_ylabel('count' , weight= 'semibold' , fontname ='monoscape' , size = 20)
axes[1].set_xlabel('Programing Language' , weight= 'semibold' , fontname ='monoscape' , size = 20)
plt.suptitle('Programimg Language Distribution of Kaggle Users' , weight= 'semibold' , fontname ='monoscape' , size = 20)
plt.show()
```
